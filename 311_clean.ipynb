```python
import pandas as pd

df = pd.read_csv("311_Service_Requests_20260620.csv")
```

```python
df.info()
```

```python
fmt = "%m/%d/%Y %I:%M:%S %p"
df["CREATED_DATE"] = pd.to_datetime(df["CREATED_DATE"], format=fmt, errors="coerce")
df["CLOSED_DATE"]  = pd.to_datetime(df["CLOSED_DATE"],  format=fmt, errors="coerce")
```

```python
df.info()
```

```python
df["RESOLUTION_DAYS"] = (df["CLOSED_DATE"] - df["CREATED_DATE"]).dt.total_seconds() / 86400
```

```python
df["RESOLUTION_DAYS"].describe()
```

```python
# Full set (70k) stays as df.

# Creating a table for resolution-time analysis: only rows that have a resolution time.
resolved = df[df["RESOLUTION_DAYS"].notna()].copy()

print("Full dataset:", len(df), "rows")
print("Resolved (for timing analysis):", len(resolved), "rows")
```

```python
df.to_csv("chicago_311_all.csv", index=False)
resolved.to_csv("chicago_311_resolved.csv", index=False)
```
